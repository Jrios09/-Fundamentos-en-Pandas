```python
##EJERCICIO A:Ejercicios de Estructura de una Gráfica
##Crear una Figura con plt.figure()
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.figure(figsize=(6,4))  # Crear figura
plt.plot(x, y)

plt.title("Gráfica usando figure()")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

plt.show()
```

```python
##Crear una Figura con plt.subplots()
fig, ax = plt.subplots()

ax.plot(x, y)
ax.set_title("Gráfica usando subplots()")
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")

plt.show()
```

```python
##Crear ejes con add_subplot()
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(x, y)
ax.set_title("Gráfica con add_subplot")

plt.show()
```

```python
##Añadir título, etiquetas, leyendas, anotaciones
plt.figure()

plt.plot(x, y, label="Datos", color='blue')

plt.title("Gráfica completa")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

plt.legend()

plt.annotate("Punto máximo", 
             xy=(4, 30), 
             xytext=(3, 25),
             arrowprops=dict(facecolor='black'))

plt.show()
```

```python
## Ejemplo con Pyplot

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("Gráfica con Pyplot")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

plt.show()
```

```python
## 1. Histograma básico

import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)

plt.hist(data)
plt.title("Histograma básico")
plt.show()
```

```python
## 2. Histograma acumulativo

plt.hist(data, cumulative=True)
plt.title("Histograma acumulativo")
plt.show()
```

```python
## 3. Histograma normalizado

plt.hist(data, density=True)
plt.title("Histograma normalizado")
plt.show()
```

```python
## 4. Boxplot básico

plt.boxplot(data)
plt.title("Boxplot básico")
plt.show()
```

```python
## 5. Boxplot personalizado

plt.boxplot(data, patch_artist=True,
            boxprops=dict(facecolor='lightblue'))
plt.title("Boxplot personalizado")
plt.show()
```

```python
## 6. Violin Plot básico

plt.violinplot(data)
plt.title("Violin Plot básico")
plt.show()
```

```python
## 7. Violin Plot personalizado

plt.violinplot(data, showmeans=True)
plt.title("Violin Plot personalizado")
plt.show()
```

```python
## 8. Gráfica de línea simple (seno)

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Función seno")
plt.show()
```

```python
## 9. Gráfica de línea personalizada

plt.plot(x, y, color='red', linestyle='--', marker='o')
plt.title("Línea personalizada")
plt.show()
```

```python
## 10. Gráfica comparativa (seno vs coseno)

plt.plot(x, np.sin(x), label='Seno')
plt.plot(x, np.cos(x), label='Coseno')

plt.legend()
plt.title("Seno vs Coseno")

plt.show()
```

```python
## 1. Scatter Plot simple

import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y)
plt.title("Scatter Plot simple")
plt.show()
```

```python
## 2. Scatter con tamaño y color

plt.scatter(x, y, s=100, c='red')
plt.title("Scatter con tamaño y color")
plt.show()
```

```python
## 3. Scatter con mapa de color

colors = np.random.rand(100)

plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.title("Scatter con mapa de color")
plt.show()
```

```python
## 4. Scatter con línea de regresión

coef = np.polyfit(x, y, 1)
poly = np.poly1d(coef)

plt.scatter(x, y)
plt.plot(x, poly(x), color='red')
plt.title("Scatter con regresión lineal")
plt.show()
```

```python
## 5. Gráfica de línea comparativa

plt.plot(x, y, label="Variable 1")
plt.plot(x, x*0.5, label="Variable 2")

plt.legend()
plt.title("Líneas comparativas")
plt.show()
```

```python
## 6. Línea con regresión cuadrática

coef = np.polyfit(x, y, 2)
poly = np.poly1d(coef)

plt.scatter(x, y)
plt.plot(x, poly(x))
plt.title("Regresión cuadrática")
plt.show()
```

```python
## 7. Heatmap básico con imshow()

matrix = np.random.rand(5,5)

plt.imshow(matrix)
plt.colorbar()
plt.title("Heatmap básico")
plt.show()
```

```python
## 8. Heatmap personalizado

plt.imshow(matrix, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.title("Heatmap personalizado")
plt.show()
```

```python
## 9. Heatmap de matriz de correlación

data = np.random.rand(10,5)
corr = np.corrcoef(data)

plt.imshow(corr, cmap='coolwarm')
plt.colorbar()
plt.title("Matriz de correlación")
plt.show()
```

```python
## 10. Heatmap con anotaciones numéricas

for i in range(len(matrix)):
    for j in range(len(matrix)):
        plt.text(j, i, f"{matrix[i,j]:.2f}", ha='center')

plt.imshow(matrix)
plt.title("Heatmap con anotaciones")
plt.show()
```

```python
## 1. Barras apiladas

import numpy as np
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C']
x = np.arange(len(labels))

y1 = [10, 20, 30]
y2 = [5, 15, 10]

plt.bar(x, y1)
plt.bar(x, y2, bottom=y1)

plt.title("Barras apiladas")
plt.show()
```

```python
## 2. Barras agrupadas

width = 0.35

plt.bar(x, y1, width)
plt.bar(x + width, y2, width)

plt.title("Barras agrupadas")
plt.show()
```

```python
## 3. Barras apiladas personalizadas

plt.bar(x, y1, color='blue', hatch='//')
plt.bar(x, y2, bottom=y1, color='orange', hatch='..')

plt.title("Barras personalizadas")
plt.show()
```

```python
## 4. Barras con etiquetas numéricas

bars = plt.bar(x, y1)

for bar in bars:
    plt.text(bar.get_x(), bar.get_height(),
             str(bar.get_height()))

plt.title("Barras con etiquetas")
plt.show()
```
